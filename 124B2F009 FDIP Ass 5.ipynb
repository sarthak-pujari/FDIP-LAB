```python
# Install OpenCV (if not already installed, usually Colab has it)
!pip install opencv-python-headless

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()   # Choose "contrast.jpeg"

# Load the uploaded image in grayscale
img = cv2.imread("Bappa.jpg", cv2.IMREAD_GRAYSCALE)

# Contrast stretching
min_val = np.min(img)
max_val = np.max(img)
contrast_stretched = ((img - min_val) / (max_val - min_val) * 255).astype(np.uint8)

# Histogram equalization
hist_equalized = cv2.equalizeHist(img)

# Display images and histograms
plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(3, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Histogram of Original
plt.subplot(3, 2, 2)
plt.hist(img.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Original")

# Contrast Stretched Image
plt.subplot(3, 2, 3)
plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched")
plt.axis("off")

# Histogram of Contrast Stretched
plt.subplot(3, 2, 4)
plt.hist(contrast_stretched.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Contrast Stretched")

# Histogram Equalized Image
plt.subplot(3, 2, 5)
plt.imshow(hist_equalized, cmap='gray')
plt.title("Histogram Equalized")
plt.axis("off")

# Histogram of Equalized Image
plt.subplot(3, 2, 6)
plt.hist(hist_equalized.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Equalized")

plt.tight_layout()
plt.show()
```
